```python
profile.to_notebook_iframe()
```

```python
train_df = df.copy()
```

```python
train_df['SalePrice'].mean()
```

```python
test_df = pd.read_csv('test.csv')
```

```python
test_df['SalePrice'] = pd.Series([train_df['SalePrice'].mean()] * len(train_df))
```

```python
test_df
```

```python
test_df[['Id', 'SalePrice']]
```

# First Try with mean prediction

```python
test_df[['Id', 'SalePrice']].to_csv('average_prediction.csv', index=False)
```

```python
!kaggle competitions submit -c house-prices-advanced-regression-techniques -f average_prediction.csv -m "average-prediction"
```

```python
from copy import deepcopy
train_df_sub = deepcopy(train_df[['MSSubClass', 'LotFrontage', 'SalePrice']])
train_df_sub
```

```python
train_df_sub.mean()
```

```python
train_df_sub = train_df_sub.fillna(train_df_sub.mean())
```

```python
X_train, y_train = train_df_sub.to_numpy()[:, :-1], train_df_sub.to_numpy()[:, -1]
```

```python
X_train
```

```python
y_train
```

# Model
- Feature and Target values - X, y

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression().fit(X_train, y_train)

lr
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
train_df_sub['Linear Prediction'] = lr.predict(X_train)
train_df_sub
```

# Evaluate Performance on Test Set

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_train, train_df_sub['Linear Prediction'])
```

```python
test_df_sub = deepcopy(test_df[['MSSubClass', 'LotFrontage']])
test_df_sub
```

```python
test_df_sub = test_df_sub.fillna(test_df_sub.mean())
test_df_sub
```

```python
X_test = test_df_sub.to_numpy()
X_test
```

```python
test_df['SalePrice'] = lr.predict(X_test)
test_df
```
